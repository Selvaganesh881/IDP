```python
from openai import OpenAI

# 1. Initialize the client to point to your local vLLM server
client = OpenAI(
    base_url="http://localhost:8000/v1", # Make sure this matches your server IP/Port
    api_key="abc-123" # Local vLLM doesn't require a real API key!
)

# 2. This is the faked text coming from your RoBERTa/Presidio pipeline
faked_kyc_document = """
Customer Name: John Doe
Address: 123 Fake Street, Faketon, TX
Account Type: Premium Business
"""

# 3. Call your open-source model exactly like you would call GPT-4
response = client.chat.completions.create(
    model="Qwen3-4B", 
    messages=[
        {
            "role": "system", 
            "content": "You are a financial AI. Extract the KYC data from the provided text and output it STRICTLY as a JSON object."
        },
        {
            "role": "user", 
            "content": faked_kyc_document
        }
    ],
    temperature=0.1 # Keep it low for deterministic JSON extraction
)

# 4. Print the extracted JSON
print(response.choices[0].message.content)
```
